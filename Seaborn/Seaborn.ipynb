```python
import matplotlib.pyplot as plt

plt.figure()
x1 = [10, 20, 5, 40, 8]
x2 = [30, 43, 9, 7, 20]

# Correcting the typo in the label argument
plt.plot(x1, label="Group 1")
plt.plot(x2, label="Group 2")

plt.legend()
plt.show()
```

```python
import seaborn as sns
sns.set()
plt.figure()
plt.plot(x1, label="Group 1")
plt.plot(x2, label="Group 2")

plt.legend()
plt.show()
```

```python
sns.set_style("whitegrid")
plt.figure()
plt.plot(x1, label="Group 1")
plt.plot(x2, label="Group 2")

plt.legend()
plt.show()
```

```python
sns.set()
plt.figure()
x1 = [10, 20, 5, 40, 8]
x2 = [30, 43, 9, 7, 20]

with sns.axes_style("dark"):
    plt.plot(x1, label="Group 1")
    plt.plot(x2, label="Group 2")
plt.legend()
plt.show()
```

```python
sns.set_context("poster")
plt.figure()
x1 = [10, 20, 5, 40, 8]
x2 = [30, 43, 9, 7, 20]

with sns.axes_style("dark"):
    plt.plot(x1, label="Group 1")
    plt.plot(x2, label="Group 2")
plt.legend()
plt.show()
```

```python
df = pd.read_csv("Air_Traffic_Passenger_Statistics.csv")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your data is stored in a DataFrame called 'df'
# If not, you can load your data using pd.read_csv() or another method

# Convert 'Year' and 'Month' columns to datetime format
df['Date'] = pd.to_datetime(df['Year'].astype(str) + '-' + df['Month'], format='%Y-%B')
df = df.sort_values(by='Date')

# Pivot the DataFrame to create a matrix suitable for a heatmap
heatmap_data = df.pivot_table(values='Passenger Count', index='Year', columns='Month', aggfunc='sum')

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='g', linewidths=.5, cbar_kws={'label': 'Passenger Count'})
plt.title('Passenger Count Heatmap by Year and Month')
plt.show()
```
